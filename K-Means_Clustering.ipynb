```
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the Iris dataset
iris = load_iris()

# 2. Extract features only (remove labels)
X = iris.data   # features: sepal length, sepal width, petal length, petal width

# 3. Create and run K-means (3 clusters because iris has 3 species)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# 4. Get cluster assignments
clusters = kmeans.labels_

# Display results
df = pd.DataFrame(X, columns=iris.feature_names)
print(df.head())
print()
df['Cluster'] = clusters

#Visulaization of clusters
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis')
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.title("K-Means Clustering on Iris Dataset")
plt.show()
```

```
import seaborn as sns

sns.pairplot(df, hue='Cluster', palette='viridis')
plt.suptitle("K-Means Clusters on Iris Dataset (Pairplot)", y=1.02)
plt.show()
```

```
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:,0], X[:,1], X[:,2], c=clusters, cmap='viridis', s=50)
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
ax.set_zlabel("Petal Length")
ax.set_title("3D K-Means Clustering on Iris Dataset")

plt.show()
```
